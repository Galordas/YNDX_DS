Let's make a new column and strip 'locality_name' of the type.

```python
# print(data['locality_name'].value_counts())
locality_type = ['поселок станции', 'посёлок', 'поселок', 'при железнодорожной станции ', 'платформа ', 'садоводческое некоммерческое товарищество ',
             'садовое товарищество ', 'станции', 'село ', 'пансионат ', 'городского типа ',
             'деревня ', 'городской поселок ']
compound_types = [('поселок станции', 'поселок'), ('посёлок станции', 'поселок'), ('посёлок городского типа', 'ПГТ'), ('поселок городского типа', 'ПГТ')]
def locality_name_change(row):
    found = False
    for compound_type in compound_types:
        if compound_type[0] in str(row['locality_name']):
            row['locality_type'] = compound_type[1].replace('ё','е').strip() # set a value to the column
            row['locality_name'] = row['locality_name'].replace('ё','е').replace(compound_type[0],'').strip()  # strip locality_name from type
            found = True
            break
    if not found:
        for loc_type in locality_type:
            if loc_type in str(row['locality_name']):
                row['locality_type'] = loc_type.replace('ё','е').strip() # set a value to the column
                row['locality_name'] = row['locality_name'].replace('ё','е').replace(loc_type,'').strip()  # strip locality_name from type
    return row

data = data.apply(locality_name_change, axis=1)
# print(data['locality_type'].unique())

# print(data['locality_name'].isna().sum())
data['locality_name'] = data['locality_name'].fillna('unknown')
data['locality_type'] = data['locality_type'].fillna('город')
print(data[['locality_name','locality_type']].tail(30))
```

Let's keep our data in the interquartile range and detect outliers in Data.
Data points that fall below `Q1−1.5×IQR` or above `Q3+1.5×IQR` are considered outliers.
* Q2 = median of the dataset.
* Q1 = median of n smallest data points.
* Q3 = median of n highest data points.

```python
columns_to_modify = ['last_price','total_area','living_area','kitchen_area','rooms','ceiling_height']
def iqr(df, columns, iqr_multiplier=1.5):
    
    # Create the mask to keep all rows initially 
    keep_mask = pd.Series(True,index=df.index)
    
    for column in columns:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1

        # Update the mask to incllude only rows within interquartile range
        column_mask = (df[column] >= q1 - iqr_multiplier * iqr) & (df[column] <= q3 + iqr_multiplier * iqr)
        keep_mask &= column_mask
    
    return df[keep_mask]   

filtered_data = iqr(data,columns_to_modify)

# We can do it in more simplier way but mask solution has following bonuses:
# * Avoid cumulative filtering - All columns are filtered simultaneously, not sequentially
# * Avoid copying dataframe multiple times

# def iqr(df,columns):
#     filtered_df = df.copy()
#     for column in columns:
#         q1 = filtered_df[column].quantile(0.25)
#         q3 = filtered_df[column].quantile(0.75)
#         iqr = q3 - q1
#         filtered_df = filtered_df.loc[(filtered_df[column] < q3+1.5*iqr) & (filtered_df[column] > q1-1.5*iqr)]
#     return filtered_df

filtered_data = iqr(data,columns_to_modify)
hard_filtered_data = iqr(data,columns_to_modify,iqr_multiplier=1)
print(filtered_data[['last_price','total_area','living_area','rooms','ceiling_height','days_exposition','city_centers_nearest']].corr())
print(hard_filtered_data[['last_price','total_area','living_area','rooms','ceiling_height','days_exposition','city_centers_nearest']].corr())
```
