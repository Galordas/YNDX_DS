#### Декомпозиция задачи
Если толокером может стать кто угодно, как добиться должного качества разметки? Сложную задачу нужно разбить на более мелкие и простые.
Декомпозиция задачи позволит:
уменьшить количество ошибок среди исполнителей;
упростить процесс проверки выполненного задания;
снизить порог входа для исполнителей: чем доступнее задача, тем больше человек выполнит её корректно и быстро.
Декомпозировать можно:
1.вертикально
* деление задачи на несколько проектов (например, прежде чем выполнить сегментацию объектов, нужно сперва определить класс изображения в датасете, а потом сегментировать объекты);
* деление проекта на несколько шагов (например, из классификации сделать мультиклассификацию).
горизонтально
постприёмка, то есть проверка выполненных другими исполнителями заданий.

2.горизонтально
* постприёмка, то есть проверка выполненных другими исполнителями заданий.
