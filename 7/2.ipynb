### Контрольные задания (golden sets)
Это задания с заведомо правильными ответами, которые заказчик добавляет в проект вперемешку с основными. Контрольные задания нужны, чтобы оценить качество работы исполнителя в режиме реального времени. Если исполнитель часто выполняет их неправильно, его доступ к проекту закрывают.

### Утечка целевого признака (target leakage)
Она происходит, когда в признаки случайно попадает информация о целевом признаке.


#### Common Causes of Target Leakage

1. Including Future Information in Features

Example: Predicting tomorrow's stock price today, but using features that incorporate tomorrow's stock price indirectly.

2. Incorrect Feature Engineering

Example: Calculating an average including the test data (before train-test split).

3. Time-Based Data Leakage

Example: Using future timestamps to predict past events.

4. Data Preprocessing Applied Before Train-Test Split

Example: Normalizing/scaling the entire dataset before splitting it.

5. Redundant Features Derived from the Target

Example: Using price_per_sqft as a feature when predicting total_price.

### Кросс-валидация (cross validation)

Как она работает? Все данные мы поделили на обучающую и тестовую выборки. Тестовую «прячем» для финальной проверки, а обучающую случайным образом делим на K равных частей. Сам метод деления называется K-Fold, где К — это число блоков (англ. folds).

Метод кросс-валидации напоминает бутстреп тем, что генерируется несколько выборок. Только в перекрёстной проверке содержимое блоков зафиксировано и не меняется на каждом этапе обучения-проверки. Каждое наблюдение проходит и через обучающий, и валидационный наборы данных.

Кросс-валидация нужна, когда нужно сравнить модели, подобрать гиперпараметры или оценить полезность признаков. Она минимизирует случайность разделения данных и даёт более точный результат. Сравните: при разовом делении мы получаем только одну валидационную выборку (25% от всех данных) и одну обучающую (75%). При перекрёстной проверке валидационный набор составляет 20%, но у нас их — пять! Значит, при каждом делении мы берём новые 20% и проверяем модель уже на 100% данных.

Единственный недостаток кросс-валидации — время на вычисления, особенно при большом количестве наблюдений или блоков. Много времени.





```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
data = pd.read_csv(r'E:\COURSES\Кодинг\[Yandex praxis]. DS specialist + (2022)\7\01 Машинное обучение в бизнесе\04 Сбор данных\heart.csv')
features = data.drop(['target'], axis=1)
target = data['target']

scores = []
# зададим размер блока, если их всего три
sample_size = int(len(data) / 3)
for i in range(0, len(data), sample_size):
    valid_indexes = data.index[i : i + sample_size]
    train_indexes = data.index.difference(valid_indexes)
    features_train = features.loc[train_indexes]
    target_train = target.loc[train_indexes]
    features_test = features.loc[valid_indexes]
    target_test = target.loc[valid_indexes]
    model = DecisionTreeClassifier(random_state=0)
    model = model.fit(features_train, target_train)
    score = model.score(features_test, target_test)
    scores.append(score)
final_score = sum(scores) / len(scores)
print('Avg score using cross validation:', final_score)

# alternatively
features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.2, random_state=0)
model = DecisionTreeClassifier(random_state=0)
model = model.fit(features_train, target_train)
score = model.score(features_valid, target_valid)
print('Avg score without cross validation', score)
```
