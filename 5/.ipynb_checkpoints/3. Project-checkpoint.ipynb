## Разбейте данные на выборки

```python
from sklearn.model_selection import train_test_split

df_train, df_valid = train_test_split(df, test_size=0.25, random_state=333)
features_train = df_train.drop('is_ultra', axis=1)
target_train = df_train['is_ultra']
features_valid = df_valid.drop('is_ultra', axis=1)
target_valid = df_valid['is_ultra']
```

## Исследуйте модели

### 1. Logistic regression

```python
from sklearn.linear_model import LogisticRegression

logistic_solvers = ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga', 'newton-cholesky']
best_result_logistic_regression = 0
best_model_logistic = None
best_solver_logistic = None

for logistic_solver in logistic_solvers:
    model = LogisticRegression(random_state=333, solver=logistic_solver, max_iter=4000)
    model.fit(features_train, target_train)
    prediction = model.predict(features_valid)
    result = model.score(features_valid, target_valid)
    print('The logistic regression solver is:', logistic_solver, 'The score is:', result)
    if result > best_result_logistic_regression:
        best_result_logistic_regression = result
        best_model_logistic = model
        best_solver_logistic = logistic_solver
print()
print('The best logistic regression solver is', best_solver_logistic, 'The best score is:', best_result_logistic_regression)
```

### 2. Random forest

```python
from sklearn.ensemble import RandomForestClassifier

best_result_random_forest = 0
best_model_random_forest = None
best_depth_random_forest = None
best_est_random_forest = None

for est in range(1, 51, 1):
    for depth in range(1, 11):
        model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
        model.fit(features_train, target_train)
        predictions = model.predict(features_valid)
        result = model.score(features_valid, target_valid)
    if result > best_result_random_forest:
        best_model_random_forest = model
        best_result_random_forest = result
        best_est_random_forest = est
        best_depth_random_forest = depth
print('The best random forest est', best_est_random_forest, 'The best random forest depth', best_depth_random_forest,
      'The best score is:', best_result_random_forest)
```

### 3. Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

best_result_decision_tree = 0
best_model_decision_tree = None
best_depth_decision_tree = None

for depth in range(1, 11):
    model = DecisionTreeClassifier(random_state=333, max_depth=depth)
    model.fit(features_train, target_train)
    predictions = model.predict(features_valid)
    result = model.score(features_valid, target_valid)
if result > best_result_decision_tree:
    best_model_decision_tree = model
    best_result_decision_tree = result
    best_depth_decision_tree = depth
print('The best decision tree depth', best_depth_decision_tree,
      'The best score is:', best_result_decision_tree)
```

## Сводные данные по работе моделей

```python
data = {
    'model': ['Decision tree', 'Random forest', 'Logistic regression'],
    'accuracy': [best_result_decision_tree, best_result_random_forest, best_result_logistic_regression]  # Замените эти значения на ваши
}
data = pd.DataFrame(data)
data
```

```python
acc_max = data['accuracy'].max()
print(f'The best result is: {acc_max} and the best model is {data.loc[data['accuracy'] == acc_max, 'model'].values[0]}')
```
