### 4.5. –õ–æ–≥–∏–∫–∞ –ø—Ä–æ–≤–µ—Ä–∫–∏ –¥–≤—É—Ö—Å—Ç–æ—Ä–æ–Ω–Ω–∏—Ö –≥–∏–ø–æ—Ç–µ–∑

–ï—Å–ª–∏ H‚ÇÄ –≤–µ—Ä–Ω–∞, –∏—Å—Ç–∏–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ —Ä–∞–≤–Ω–æ —á–∏—Å–ª—É, –∫–æ—Ç–æ—Ä–æ–µ –≤ —ç—Ç–æ–π –≥–∏–ø–æ—Ç–µ–∑–µ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏–ª–∏. –°–æ–≥–ª–∞—Å–Ω–æ –¶–ü–¢, –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ —Å—Ä–µ–¥–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ –≤–æ–∫—Ä—É–≥ –∏—Å—Ç–∏–Ω–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏.

–¢–µ–ø–µ—Ä—å –º–æ–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å, –∫—É–¥–∞ –Ω–∞ —ç—Ç–æ–º –≥—Ä–∞—Ñ–∏–∫–µ –ø–æ–ø–∞–ª–æ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ ‚Äî —á–∏—Å–ª–æ, –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ –Ω–∞ —Ç–æ–π –≤—ã–±–æ—Ä–∫–µ, –∫–æ—Ç–æ—Ä–∞—è –µ—Å—Ç—å –≤ –Ω–∞—à–µ–º —Ä–∞—Å–ø–æ—Ä—è–∂–µ–Ω–∏–∏. –ü—Ä–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞—Å—Ç—å –≤ —Ç–æ—Ç –∏–ª–∏ –∏–Ω–æ–π –∏–Ω—Ç–µ—Ä–≤–∞–ª —Ä–∞–≤–Ω–∞ –ø–ª–æ—â–∞–¥–∏ –≥—Ä–∞—Ñ–∏–∫–∞ –Ω–∞–¥ —ç—Ç–∏–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–º. –í —Ä–∞–π–æ–Ω–µ —Ü–µ–Ω—Ç—Ä–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –≤ –Ω–µ–∫–æ—Ç–æ—Ä–æ–º –¥–∏–∞–ø–∞–∑–æ–Ω–µ –≤–æ–∫—Ä—É–≥ –Ω–µ–≥–æ –±—É–¥—É—Ç –≤–µ—Ä–æ—è—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è. 

¬´–£–µ—Ö–∞—Ç—å¬ª –≤–ø—Ä–∞–≤–æ-–≤–ª–µ–≤–æ –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –Ω–∞ –æ–¥–Ω—É –∏–ª–∏ –¥–∞–∂–µ –¥–≤–µ —Å–∏–≥–º—ã ‚Äî –æ–±—ã—á–Ω–æ–µ –¥–µ–ª–æ. –≠—Ç–æ –µ—â—ë –Ω–µ –ø–æ–≤–æ–¥ –æ—Ç–≤–µ—Ä–≥–∞—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É. (+-2.5%)

–ú–æ–∂–Ω–æ –∑–∞–¥–∞—Ç—å –∏ –¥—Ä—É–≥–æ–π —É—Ä–æ–≤–µ–Ω—å —á–µ—Ä–µ–∑ –∫–æ–º–∞–Ω–¥—É `alpha = .05` (—É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏) - significance level for the hypothesis test

```python
from scipy import stats as st 
import numpy as np
np.random.seed(88)
sample = np.random.normal(1000, 25, 100) # mean 1000, 20 - standard deviation, 30 - amount of numbers

result = st.ttest_1samp(sample, 999)
print(sample)
print(result)
```
